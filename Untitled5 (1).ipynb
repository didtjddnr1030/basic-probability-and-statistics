```python
import matplotlib.pyplot as plt
joint_pmf_data={
    (1, 1):0.1,
    (1, 2):0.2,
    (1, 3):0.2,
    (2, 1):0.3,
    (2, 2):0.05,
    (2, 3):0.15
}
f_X={}
f_Y={}
for x,y in joint_pmf_data.keys():
  f_X[x]=f_X.get(x,0)+joint_pmf_data[(x,y)]
for x,y in joint_pmf_data.keys():
  f_Y[y]=f_Y.get(y,0)+joint_pmf_data[(x,y)]
plt.figure(figsize=(12,5))
y_min=0
y_max=1
x_min=0.5
x_max=3.5
plt.subplot(1,2,1)
plt.plot(list(f_X.keys()), list(f_X.values()),'bo')
plt.vlines(list(f_X.keys()),y_min,
list(f_X.values()),colors='blue')
plt.title('Marginal PMF of X')
plt.xlabel('X value')
plt.ylabel('Probability')
plt.xticks(range(int(x_min), int(x_max)+1))
plt.xlim(x_min,x_max)
plt.ylim(y_min,y_max)
plt.subplot(1,2,2)
plt.plot(list(f_Y.keys()),list(f_Y.values()),'ro')
plt.vlines(list(f_Y.keys()),y_min,list(f_Y.values()), colors='red')
plt.title('Marginal PMF ofY')
plt.xlabel('Y value')
plt.ylabel('Probability')
plt.xticks(range(int(x_min),int(x_max)+1))
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.tight_layout()
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x=np.array([3,2,0,1,2,4,1,2,1,3,2,0,3,1,0])
y=np.array([15,21,25,23,19,20,23,19,20,17,21,28,20,22,26])
cov_xy=np.average((x-np.average(x))*(y-np.average(y)))
print("cov_xy", cov_xy)
plt.scatter(x,y,marker="o", label="xy", s=40)
plt.legend()
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```
