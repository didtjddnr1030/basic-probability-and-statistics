```python
import math

# 노드 클래스 (원형 리스트용)
class Node:
    def __init__(self, name, x, y):
        self.name = name
        self.x = x
        self.y = y
        self.link = None

# 거리 계산
def dist(x1, y1, x2, y2):
    dx = x1 - x2
    dy = y1 - y2
    return (dx*dx + dy*dy) ** 0.5

# 전체 경로 거리 계산
def totaldist(route):
    total = 0
    # 모항 -> 첫 항구
    total += dist(0, 0, route[0][1], route[0][2])

    # 항구들 사이 이동
    for i in range(len(route)-1):
        total += dist(route[i][1], route[i][2], route[i+1][1], route[i+1][2])

    # 마지막 항구 -> 모항
    total += dist(route[-1][1], route[-1][2], 0, 0)
    return total

# 가능한 경로 중 최단 경로 찾기
def shortroute(portList):
    n = len(portList)
    best = None
    mind = 999999999

    if n == 1:
        return portList, totaldist(portList)

    # n개에 따라 모든 순열을 직접 반복문으로 탐색
    if n == 2:
        for i in range(n):
            for j in range(n):
                if i != j:
                    r = [portList[i], portList[j]]
                    d = totaldist(r)
                    if d < mind:
                        mind = d
                        best = r

    elif n == 3:
        for i in range(n):
            for j in range(n):
                for k in range(n):
                    if len({i, j, k}) == 3:
                        r = [portList[i], portList[j], portList[k]]
                        d = totaldist(r)
                        if d < mind:
                            mind = d
                            best = r

    elif n == 4:
        for i in range(n):
            for j in range(n):
                for k in range(n):
                    for l in range(n):
                        if len({i, j, k, l}) == 4:
                            r = [portList[i], portList[j], portList[k], portList[l]]
                            d = totaldist(r)
                            if d < mind:
                                mind = d
                                best = r

    elif n == 5:
        for i in range(n):
            for j in range(n):
                for k in range(n):
                    for l in range(n):
                        for m in range(n):
                            if len({i, j, k, l, m}) == 5:
                                r = [portList[i], portList[j], portList[k], portList[l], portList[m]]
                                d = totaldist(r)
                                if d < mind:
                                    mind = d
                                    best = r

    return best, mind

# 원형 리스트 만들기
def circle(route):
    head = None
    prev = None
    for name, x, y in route:
        node = Node(name, x, y)
        if head is None:
            head = node
            prev = node
        else:
            prev.link = node
            prev = node
    prev.link = head
    return head

# 경로 출력
def printroute(route):
    print("\n 최단 항로 출력")
    print("모항(0,0)")

    total = 0

    d = dist(0, 0, route[0][1], route[0][2])
    print(f" ↓ {d:.2f} km")
    total += d
    print(f"{route[0][0]}({route[0][1]}, {route[0][2]})")

    for i in range(len(route)-1):
        d = dist(route[i][1], route[i][2], route[i+1][1], route[i+1][2])
        print(f" ↓ {d:.2f} km")
        total += d
        print(f"{route[i+1][0]}({route[i+1][1]}, {route[i+1][2]})")

    d = dist(route[-1][1], route[-1][2], 0, 0)
    print(f" ↓ {d:.2f} km")
    total += d
    print("모항(0,0)")

    print(f"\n총 이동거리: {total:.2f} km")

# 메인
if __name__ == "__main__":
    ports = []
    print("항구 입력 (최대 5개)")

    while True:
        name = input("항구 이름 (엔터 시 종료): ")
        if name == "":
            break
        x = int(input("x 좌표: "))
        y = int(input("y 좌표: "))
        ports.append([name, x, y])

        if len(ports) == 5:
            print("최대 5개까지 입력 가능.")
            break

    if len(ports) == 0:
        print("입력된 항구 없음.")
        exit()

    best, md = shortroute(ports)
    head = circle(best)
    printroute(best)
```
