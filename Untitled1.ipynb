```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5,6,7])#날짜 데이터
y=np.array([78.4,78.6,78.5,78.9,78.2,79.4,78.6])#체중 데이터
plt.plot(x,y,color="r")#빨간색 선으로 그래프를 나타낸다
plt.show()#그래프를 보여준다
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5,6,7])
y=np.array([78.4,78.6,78.5,78.9,78.2,79.4,78.6])
plt.scatter(x,y,color="r")#빨간색으로 산포도 형식으로 보여준다
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5,6,7])
y=np.array([78.4,78.6,78.5,78.9,78.2,79.4,78.6])
plt.plot(x,y,color="r",marker="o")#색은 빨간색, 점을 추가해서 그래프로 나타낸다
plt.grid()#그리드(격자) 형식으로 나타낸다
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(-5,5,300)#-5에서 5까지 300개의 구간으로 나눈다
yl=x**2
y2=(x-2)**2
plt.plot(x,yl,color="r")#선은 빨간색
plt.plot(x,y2,color="k",linestyle="--")#색은 검정색, 선 스타일은 점선
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np
font_path="/usr/share/fonts/truetype/nanum/NanumGothic.ttf"
fm.fontManager.addfont(font_path)
plt.rc('font',family=fm.FontProperties(fname=font_path).get_name())
plt.rcParams["axes.unicode_minus"]=False
x=np.linspace(-5,5)
y1=2*x
y2=5*x
plt.plot(x,y1,label="yl값", color='r')#빨간 선
plt.plot(x,y2,label="y2값", linestyle="dashed", color='k')#검정 점선
plt.xlabel("X-축", size=14)
plt.ylabel("Y=축", size=14)
plt.title("그래프 제목")
plt.grid()
plt.legend()
plt.show()
```
